```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://zestrogaming.com/product-category/graphics-card-price-in-pakistan/"
request = requests.get(url)
request
```

```python
# request.content
```

```python
soup = BeautifulSoup(request.content, 'html.parser')
# soup
```

```python
graphicCards = soup.find_all(class_='jet-woo-products__item jet-woo-builder-product')
graphicCardsData = []

for card in graphicCards:
    # Get the product title
    title = card.find(class_='jet-woo-product-title').text.strip() if card.find(class_='jet-woo-product-title') else 'No Title'
    
    # Get the product price, handle missing price gracefully
    price_element = card.find(class_='jet-woo-product-price')
    price = price_element.text.strip().split()[-1] if price_element else 'No Price'
    
    # Get the product info (e.g., model or description)
    info = card.find(class_='jet-woo-product-excerpt').text.strip() if card.find(class_='jet-woo-product-excerpt') else 'No Info'

    # Assuming 'info' contains <ul><li> elements for the model details
    model = None  # Initialize model variable
    features = []  # List to store additional features (other <li> elements)

    # Look for <ul> and <li> tags for model and features
    ul_tag = card.find('ul')  # Assuming the <ul> tag contains <li> elements
    if ul_tag:
        lis = ul_tag.find_all('li')  # Find all <li> elements
        for li in lis:
            if "model" in li.text.lower():  # Look for "model" in the text
                model = li.text.strip()  # Assign model if found
            else:
                features.append(li.text.strip())  # Add other li items to features

    # If model is not found, assign a default value
    model = model if model else "Model not found"

    # Store the extracted data (title, price, model, and features)
    graphicCardsData.append({"title":title, "price": price, "model": model, "features" :features})

# Print the collected data
graphicCardsData
```

```python
import os
import csv

# Define your relative path to the target folder
folder_path = os.path.join('..', 'data', 'scraped')  # '..' means going one level up to 'scraper'

# Define the full file path
file_path = os.path.join(folder_path, 'zestro.csv')

# Make sure the directory exists
if not os.path.exists(folder_path):
    os.makedirs(folder_path)  # This ensures the directory exists, though you said it's already made

# Define the header (column names for the CSV file)
header = ["title", "price", "model", "features"]

# Open a CSV file to write the data
with open(file_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=header)
    
    # Write the header to the CSV file
    writer.writeheader()
    
    # Write each row of data (from graphicCardsData)
    for card_data in graphicCardsData:
        writer.writerow(card_data)

print(f"Data has been written to {file_path}")
```
